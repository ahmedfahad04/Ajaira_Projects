```python
import os
import zipfile
import shutil
import pandas as pd

try:
	root_dir = os.path.dirname(os.path.dirname(os.path.abspath(__file__)))
except NameError:
	root_dir = os.path.dirname(os.path.dirname(os.getcwd()))

part_1_dir = os.path.join(root_dir, "output", "part_1")
final_dir = os.path.join(root_dir, "output", "final")

datasets = {
	"human_eval_164": {
		"zip": "human_eval_164_gpt-5-mini_fixed_prompt_v1.zip",
		"code1": "variants_code_human_eval_164_copilot_gpt-5-mini.csv",
		"labels1": "variants_labels_human_eval_164_copilot_gpt-5-mini.csv",
		"code2": "variants_code_human_eval_164_copilot_gpt-5-mini_fixed_prompt_v1.csv",
		"labels2": "variants_labels_human_eval_164_copilot_gpt-5-mini_fixed_prompt_v1.csv",
		"variant_folder": os.path.join("scripts/output/variants", "human_eval_164_gpt-5-mini")
	},
	"classEval": {
		"zip": "classEval_gpt-5-mini_fixed_prompt_v1.zip",
		"code1": "variants_code_classEval_copilot_gpt-5-mini.csv",
		"labels1": "variants_labels_classEval_copilot_gpt-5-mini.csv",
		"code2": "variants_code_classEval_copilot_gpt-5-mini_fixed_prompt_v1.csv",
		"labels2": "variants_labels_classEval_copilot_gpt-5-mini_fixed_prompt_v1.csv",
		"variant_folder": os.path.join("scripts/output/variants", "classEval_gpt-5-mini")
	}
}

for name, cfg in datasets.items():
	print(f"
{'='*60}")
	print(f"Processing: {name}")
	print(f"{'='*60}")
	
	ds_final_dir = os.path.join(final_dir, name)
	os.makedirs(ds_final_dir, exist_ok=True)
	
	# Merge code CSVs from both sources (scripts/output + part_1)
	code1 = pd.read_csv(os.path.join(root_dir, "scripts/output", cfg["code1"]))
	code2 = pd.read_csv(os.path.join(part_1_dir, cfg["code2"]))
	merged_code = pd.concat([code1, code2], ignore_index=True)
	merged_code.drop_duplicates(subset=["task_id"], keep="first", inplace=True)
	merged_code.to_csv(os.path.join(ds_final_dir, f"variants_code_{name}_merged.csv"), index=False)
	print(f"  Code CSV: {len(merged_code)} rows")
	
	# Merge labels CSVs from both sources (includes probabilities)
	labels1 = pd.read_csv(os.path.join(root_dir, "scripts/output", cfg["labels1"]))
	labels2 = pd.read_csv(os.path.join(part_1_dir, cfg["labels2"]))
	merged_labels = pd.concat([labels1, labels2], ignore_index=True)
	merged_labels.drop_duplicates(subset=["task_id"], keep="first", inplace=True)
	merged_labels.to_csv(os.path.join(ds_final_dir, f"variants_labels_{name}_merged.csv"), index=False)
	print(f"  Labels CSV: {len(merged_labels)} rows (probabilities intact)")
	
	# Copy .py files from variants folder (highest priority)
	variants_dir = os.path.join(root_dir, cfg["variant_folder"])
	if os.path.exists(variants_dir):
		for f in [f for f in os.listdir(variants_dir) if f.endswith(".py")]:
			shutil.copy(os.path.join(variants_dir, f), os.path.join(ds_final_dir, f))
		print(f"  Copied .py from variants folder")
	
	# Extract .py from zip, strip _classification, fill gaps only
	with zipfile.ZipFile(os.path.join(part_1_dir, cfg["zip"]), 'r') as z:
		for f in z.namelist():
			if (f.endswith('.py') or f.endswith('.txt')) and '_variant' in f:
				new_name = os.path.basename(f).replace('_classification', '').rsplit('.', 1)[0] + '.py'
				dest = os.path.join(ds_final_dir, new_name)
				if not os.path.exists(dest):
					with z.open(f) as src, open(dest, 'wb') as dst:
						shutil.copyfileobj(src, dst)
	
	final_count = len([f for f in os.listdir(ds_final_dir) if f.endswith(".py")])
	print(f"  Total .py files: {final_count}")
	print(f"  Final folder: {ds_final_dir}")

print("
Done!")
```
